```python
def make_plot(df_peil, df_hydro, df_streef, df_crossings):
    plt.close("all")
    fig, ax = plt.subplots()

    old_len = len(df_peil)
    dfp = df_peil.copy().explode(index_parts=True)
    dfs = df_streef.set_index("globalid", inplace=False)
    for i, row in enumerate(dfp.itertuples()):
        coords = row.geometry.exterior.coords
        #         if old_len == len(dfp):
        xtext = row.geometry.centroid.x - 0.5
        ytext = row.geometry.centroid.y
        #         else:
        #             xtext = coords[0][0] + 0.05
        #             ytext = coords[0][1]
        if (i % 2) == 0:
            ytext += 0.05
        else:
            ytext -= 0.05

        ax.text(
            xtext,
            ytext,
            f"{row.globalid}, wl={dfs.waterhoogte.at[row.globalid]}m",
            alpha=0.5,
        )
        ax.add_patch(
            Polygon(
                coords, alpha=0.5, lw=1, facecolor="powderblue", edgecolor="skyblue"
            )
        )

    for row in df_hydro.itertuples():
        x, y = zip(*row.geometry.coords)
        offset = row.geometry.interpolate(0.1, normalized=True)
        ax.arrow(
            x[0],
            y[0],
            offset.x - x[0],
            offset.y - y[0],
            shape="full",
            lw=0,
            length_includes_head=True,
            head_width=0.05,
            color="steelblue",
        )
        ax.text(offset.x, offset.y + 0.05, row.globalid)
        ax.plot(x, y, marker=".", markersize=5, lw=1, color="steelblue")

    for row in df_crossings.itertuples():
        if row.crossing_type == 1:
            ax.plot(
                row.geometry.x,
                row.geometry.y,
                marker="s",
                markersize=5,
                color="olivedrab",
            )
        elif row.crossing_type == 2:
            ax.plot(
                row.geometry.x,
                row.geometry.y,
                marker="s",
                markersize=5,
                color="indianred",
            )
        else:
            raise ValueError(f"{row.crossing_type=} not implemented")

    ax.set_xlim([-0.1, 4.1])
    ax.set_ylim([-0.1, 2.1])
    ax.set_aspect("equal")
    fig.tight_layout()
    plt.show()
```

```python
for i, gpkg_path in enumerate(sorted(testdir.iterdir())):
    if gpkg_path.is_file() and gpkg_path.suffix == ".gpkg":
        #         if "butterfly 3" not in gpkg_path.stem:
        #             continue
        #         if "perfect fit on edge" not in gpkg_path.stem:
        #             continue
        #         if "polygon within polygon 1" not in gpkg_path.stem:
        #             continue

        display(HTML(f"<h2>Test {i+1:02d}: {gpkg_path.stem}</h2>"))

        cross = ParseCrossings(gpkg_path, disable_progress=True)
        df_crossings = cross.find_crossings_with_peilgebieden(
            "hydroobject", group_stacked=True, filterlayer=None
        )
        df_crossings = cross.add_waterlevels_to_crossings(df_crossings)
        df_crossings_valid = df_crossings[df_crossings.in_use].copy()

        display(df_crossings)
        display(df_crossings_valid)
        make_plot(
            cross.df_gpkg["peilgebied"],
            cross.df_gpkg["hydroobject"],
            cross.df_gpkg["streefpeil"],
            df_crossings_valid,
        )
        display(HTML("<hr />"))
```
